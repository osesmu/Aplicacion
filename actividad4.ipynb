```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

# actividad 4

Importamos nuestra base de datos

```python
weather =    pd.read_csv('london2018.csv')
print(weather)
```

# actividad 5

Muestra la mediana de Fisica.

```python
media = weather["Fisica"].mean()
print (media)
```

# actividad 6

```python
weather.plot.bar(y='Mates', x='Nombre')
```

Comparación de diferentes asignaturas.

```python
weather.plot(kind='bar', y=['Fisica', 'Quimica','Mates','Biologia'], subplots=True, layout=(2,2), figsize=(10,5))
plt.show()
```

Comparacion de asignaturas con otra forma

```python
weather.plot(y=['Fisica', 'Quimica','Mates','Biologia'], subplots=True, layout=(2,2), figsize=(10,5))
plt.show()
```

Podemos comprobar la mediana, y percentiles con la boxplot.

```python
weather.plot.box(y='Mates')
plt.show()
```
